```python
import pandas as pd
import matplotlib as mpl
%matplotlib inline
from statsmodels.tsa.seasonal import seasonal_decompose, STL
```

```python
import pandas_datareader as pdr

import requests_cache
session = requests_cache.CachedSession(cache_name='stocks-cache', backend='sqlite')

ts = pdr.DataReader(
    'RUB=X',
    start='2014-01-01',
    end='2020-05-01',
    data_source='yahoo',
    session=session
).Close

ts.loc[ts < 1] = None

ts = ts.fillna(method='ffill').rename('RUB=X')

ts.index = pd.to_datetime(ts.index)

dec = seasonal_decompose(ts, model='additive', freq=365)
with mpl.rc_context() as rc:
    mpl.rc("figure", figsize=(12,10))
    dec.plot();
```

```python
dec = STL(ts, period=365).fit()
with mpl.rc_context() as rc:
    mpl.rc("figure", figsize=(12,10))
    dec.plot();
```
