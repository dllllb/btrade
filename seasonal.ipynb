```python
import pandas as pd
import matplotlib as mpl
%matplotlib inline
from statsmodels.tsa.seasonal import seasonal_decompose, STL
```

```python
ts = pd.read_csv('data/RUB=X.yahoo.csv', index_col='Date').Close.fillna(method='ffill').rename('RUB=X')
ts.index = pd.to_datetime(ts.index)

dec = seasonal_decompose(ts, model='additive', freq=365)
with mpl.rc_context() as rc:
    mpl.rc("figure", figsize=(12,10))
    dec.plot();
```

```python
dec = STL(ts, period=365).fit()
with mpl.rc_context() as rc:
    mpl.rc("figure", figsize=(12,10))
    dec.plot();
```
